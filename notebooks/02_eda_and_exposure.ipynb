# 02 · Exposure of the Spanish workforce + headline figures
Merge crosswalked exposure with EPA employment and EES wages, then make the
two money charts.

```python
%load_ext autoreload
%autoreload 2
import sys; sys.path.append('..')
import pandas as pd, matplotlib.pyplot as plt
from src import data_loading as dl, config
from src.exposure import build_panel, share_of_workforce_exposed
```

```python
exposure_cno = pd.read_parquet(config.DATA_PROCESSED / 'exposure_cno.parquet')
epa = dl.load_epa('epa_2024t4.sav')
earnings = dl.load_earnings()
panel = build_panel(exposure_cno, epa, earnings)
print(f"Share of workers in high-exposure occupations: {share_of_workforce_exposed(panel):.1%}")
panel.head()
```

```python
# Money chart: wage vs exposure, bubble = employment
fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(panel['mean_wage'], panel['exposure'], s=panel['employment_share']*4000, alpha=.5)
ax.set_xlabel('Mean annual wage (EUR)'); ax.set_ylabel('AI exposure'); ax.set_title('AI exposure vs. wage — Spain')
fig.tight_layout(); fig.savefig(config.FIGURES / 'wage_vs_exposure.png', dpi=150)
```
